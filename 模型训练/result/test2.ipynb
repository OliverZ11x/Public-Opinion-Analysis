```python
import pandas as pd

# 读取VOC标签体系
voc_labels_path = r'F:\practice code\guasscode\practice_code_guasscode\舆情分析\datasets\VOC标签体系0711.xlsx'
voc_labels = pd.read_excel(voc_labels_path)

# 展示数据的前几行
voc_labels
```

```python
import json

# 打开并读取 result.json 文件
with open('result.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
# 提取 text_a 和对应的最高 score 的 label
processed_data = []
for item in data:
    text = item['text_a']
    predictions = item['predictions']
    # 检查 predictions 是否为空
    if predictions:
        # 找到最高 score 的 label
        highest_score_label = max(predictions, key=lambda x: x['score'])['label']
        processed_data.append({'text_a': text, 'label': highest_score_label})
    else:
        # 处理空 predictions 的情况，例如跳过或设置默认值
        # result.append({'text_a': text, 'label': '无预测'})
        pass

# 打印结果
for entry in processed_data:
    print(f"text_a: {entry['text_a']}, label: {entry['label']}")
```

```python
根据processed_data中数据的label匹配voc_labels的一级指标和二级指标和分类
```

```python
import pandas as pd
# 保存匹配结果的列表
results = []

# 匹配并保存结果
for entry in processed_data:
    label = entry['label']
    match = voc_labels[voc_labels.apply(lambda row: label in row.values, axis=1)].head(1)
    if not match.empty:
        match = match.iloc[0]
        results.append({
            "text": entry['text_a'],
            "分类": match['分类'],
            "一级指标": match['一级指标'],
            "二级指标": match['二级指标'],
            "三级指标": entry['label']
        })
    else:
        results.append({
            "text": entry['text_a'],
            "label": label,
            "分类": None,
            "一级指标": None,
            "二级指标": None,
            "三级指标": None
        })

# 转换为 DataFrame
results_df = pd.DataFrame(results)
results_df
```

```python
results_df.to_csv('results.csv', index=False, encoding='utf-8-sig')
```
