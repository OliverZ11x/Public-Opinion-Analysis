```python
import pandas as pd

# 读取VOC标签体系
voc_labels_path = r'F:\practice code\guasscode\practice_code_guasscode\舆情分析\datasets\VOC标签体系0711.xlsx'
voc_labels = pd.read_excel(voc_labels_path)

# 展示数据的前几行
voc_labels.head()
```

```python
# 创建字典映射
mapping_dict = {}

# 遍历数据行
for index, row in voc_labels.iterrows():
    key = row['三级指标']
    value = row['四级指标']
    if key in mapping_dict:
        mapping_dict[key].append(value)
    else:
        mapping_dict[key] = [value]

# 打印字典映射
print(mapping_dict)
```

```python
import json

# 打开并读取 result.json 文件
with open('result.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
data
```

```python
# 提取 text_a 和对应的最高 score 的 label
processed_data = []
for item in data:
    text = item['text_a']
    predictions = item['predictions']
    # 检查 predictions 是否为空
    if predictions:
        # 找到最高 score 的 label
        highest_score_label = max(predictions, key=lambda x: x['score'])['label']
        processed_data.append({'text_a': text, 'label': highest_score_label})
    else:
        # 处理空 predictions 的情况，例如跳过或设置默认值
        # result.append({'text_a': text, 'label': '无预测'})
        pass

# 打印结果
for entry in processed_data:
    print(f"text_a: {entry['text_a']}, label: {entry['label']}")
```
